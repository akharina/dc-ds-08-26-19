##### How would you pick columns 0,1,3 ?

```python
df1 = df.loc[:,[0,1,3]]
```

```python
import numpy as np
```

##### Use a for loop to find all rows where column 3 entries have values greater than 2 and replace them with NaN values 
##### Hint: Use NumPy

```python
df1.info()
```

```python
df1.iloc[0,2]
```

```python
counter = 0
for x in range(len(df1)): 
    if df1.iloc[x,2] > 2:
        print(f"Changing row {x+1} to NaN")
        df1.iloc[x,2] = np.nan
        counter += 1

print(f"Changed {counter} cells to NaN")
```

```python
df1.info()
```

##### Change names of columns to A, B, and C

```python
df1.columns = ["A", "B", "C"]
df1.info()
```

##### Add a column 'Day' and start a count of number from 1-100 for all entries

```python
df1["Day"] = np.nan

day = 1

for x in range(len(df1)):
    df1.iloc[x,3] = day
    if day == 100:
        day = 0
    day += 1    
```

```python
df1.Day.value_counts()
```

Now use a Pandas Groupby function to group by the 'Day' column and find the mean of each row

```python
df1.groupby(df1.Day).mean()
```
